```python
# Importing all the necessary libraries
import pandas as pd
import datetime as dt
from datetime import datetime, timedelta
from matplotlib import pyplot as plt  
#from matplotlib import dates as mpl_dates
```

```python
plt.style.use('default') # Defining the style of plot to be used while making the plot.
# Other options available are "seaborn","Grayscale","ggplot", "Classic" etc.
```

```python
# First I created a text file "dieselpriceorlen" containing the data of diesel prices
# of Orlen for the year 2022. Data was taken from the website 
# https://www.orlen.pl/en/for-business/fuel-wholesale-prices#fuels-archive.

data = pd.read_csv("dieselpriceorlen.txt", delimiter=",") #Allows us to read the text file into a python pandas dataframe.
#The delimeter is the one that separates the datas in the text file. 

data['Date'] = pd.to_datetime(data['Date'])
data.sort_values('Date', inplace=True)
```

```python
# From the Orlen website (https://www.orlen.pl/en/for-business/fuel-wholesale-prices#fuels-archive)
# I collected the data of 5 fuel for the year 2022 and plotted the price 
price_date = data['Date']
price_f1= data['Pricefuel1']
price_f2= data['Pricefuel2']
price_f3= data['Pricefuel3']
price_f4= data['Pricefuel4']
price_f5= data['Pricefuel5']
```

```python
p1=plt.plot(price_date,price_f1, label="Diesel Oil Arctic Grade A2")
p2=plt.plot(price_date,price_f2, label="Ekodiesel diesel oil")
p3= plt.plot(price_date,price_f3, label="Ekoterm fuel oil")
p4= plt.plot(price_date,price_f4, label="Unleaded Eurosuper 95")
p5= plt.plot(price_date,price_f5, label="Unleaded SuperPlus 98")
plt.xlabel('Date')
plt.ylabel('Price (PLN/m^3)')
plt.legend(loc="best")
plt.title("Fuel wholesale prices Orlen for the year 2022")
```

```python
# From the Orlen website (https://www.orlen.pl/en/for-business/fuel-wholesale-prices#fuels-archive)
# I collected the data of 5 fuel for the year 2021 and plotted the price 

data2021 = pd.read_csv("dieselpriceorlen2021.txt", delimiter=",") #Allows us to read the text file into a python pandas dataframe.
#The delimeter is the one that separates the datas in the text file. 

data2021['Date'] = pd.to_datetime(data2021['Date'])
data2021.sort_values('Date', inplace=True)

price_date21 = data2021['Date']
price_f121= data2021['Fuelprice1']
price_f221= data2021['Fuelprice2']
price_f321= data2021['Fuelprice3']
price_f421= data2021['Fuelprice4']
price_f521= data2021['Fuelprice5']

plt.plot(price_date21,price_f121, label="Diesel Oil Arctic Grade A2")
plt.plot(price_date21,price_f221, label="Ekodiesel diesel oil")
plt.plot(price_date21,price_f321, label="Ekoterm fuel oil")
plt.plot(price_date21,price_f421, label="Unleaded Eurosuper 95")
plt.plot(price_date21,price_f521, label="Unleaded SuperPlus 98")
plt.xlabel('Date')
plt.ylabel('Price (PLN/m^3)')
plt.legend(loc="best")
plt.title("Fuel wholesale prices Orlen for the year 2021")
```

```python
# From the Orlen website (https://www.orlen.pl/en/for-business/fuel-wholesale-prices#fuels-archive)
# I collected all the data of Diesel Oil arctic Grade A2 fuel for the year 2010-2022 and plotted the price 
#for this entire period
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
datayr = pd.read_csv("dieseloilarcticgrade2022_2010.txt", delimiter=",") #Allows us to read the text file into a python pandas dataframe.
#The delimeter is the one that separates the datas in the text file. 
datayr['Date'] = pd.to_datetime(datayr['Date'])

datayr.sort_values('Date', inplace=True)

price_dateyr = datayr['Date']
price= datayr['Price']

plt.plot(price_dateyr,price)
plt.xlabel('Date')
plt.ylabel('Price (PLN/m^3)')
plt.title("Diesel Oil Arctic Grade 2 fuel wholesale prices Orlen from the year 2010-2022")
```
