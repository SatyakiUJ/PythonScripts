```python
#Exercise 9.1)Plot functions sin(x), cos(x), and exp(-x) in a range [0,10].
#Colors are red, green, and blue, respectively. Lines are solid, dashed, and dotted, respectively. Add a legend. 
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,10,100)
p1=np.sin(x)
p2=np.cos(x)
p3=np.exp(-x)
plt.plot(x,p1,'r',label='Sin(x)')
plt.plot(x,p2,'g--',label='Cos(x)')
plt.plot(x,p3,'b:',label='Exp(-x)')
plt.legend()
```

```python
#Exercise 9.2)Generate n=100 random points in a unit square [0,1]x[0,1].
#Points are green if the distance from (0,0) is less then 1; they are red otherwise.
#The marker area of a point (x,y) should be proportional to x+y.

n = 100
x = np.random.rand(n)
y = np.random.rand(n)
colors=np.where(np.sqrt(x**2+y**2)<1, 'g', 'r')
area=x+y
plt.scatter(x, y,s=area,c=colors)
           
```
