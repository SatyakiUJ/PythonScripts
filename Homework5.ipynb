```python
# Exercise 5.1
import os

directory=os.getcwd()
print(directory)
def find_pdf_size(top): 
    for root, dirs, files in os.walk(top):
        n_pdf = 0
        for name in files:
            if name.endswith(".pdf"):
                n_pdf += 1
                print("The number of PDF files in {} is {}".format(top, n_pdf))
                print("PDF files consume {} bytes",os.path.getsize(os.path.join(root, name)))
               
    
top=os.getcwd()
print(find_pdf_size(top))
```

```python
# Exercise 5.2
from datetime import timedelta, datetime

def print_working_date(date1, date2):
    excluded = (6,7)  #To exclude Saturday and Sunday.
    for j in range(int((date2 - date1).days)+1):
        if (date1 + timedelta(j)).isoweekday() in excluded:
            continue
        else: 
            yield date1 + timedelta(j)
        
date1 = datetime.now()
date2 = datetime(2022, 4, 14) 
for dt in print_working_date(date1, date2):
    print(dt.strftime("%Y-%m-%d"))
```

```python
# Exercise 5.3
import random
num=5 #selected number of experiments.
for j in range(num): # for controlling the number of experiments.
    def random_walk(start):
        x = 0
        initial = x
        xposition = [initial]
        probabilities = [-1, 1]
        for i in range(1, start + 1):
            x += random.choice(probabilities)
            xposition.append(x)
        return xposition
    print("Result of experiment number",j+1,"=", random_walk(100))
```
